```python
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
import spacy
import re
nlp = spacy.load("en_core_web_sm")
```

```python
lemma = WordNetLemmatizer()
```

```python
#stopwords of spacy
list1 = nlp.Defaults.stop_words
print(len(list1))

#stopwords of nltk
list2 = stopwords.words('english')
print(len(list2))

#combining the stopword lists
def Union(lst1, lst2): 
    final_list = list(set(lst1) | set(lst2)) 
    return final_list 
#stopwords =set((set[list1]|set[list2]))
stopwords = Union(list1, list2)

print(len(stopwords))
```

```python
#Cleaning of text
def clean_text(text):
    
    """
    It takes text as an input and clean it by applying several methods
    
    """
    
    string = ""
    
    #lower casing
    text=text.lower()
    
    #simplifying text
    text=re.sub(r"i'm","i am",text)
    text=re.sub(r"he's","he is",text)
    text=re.sub(r"she's","she is",text)
    text=re.sub(r"that's","that is",text)
    text=re.sub(r"what's","what is",text)
    text=re.sub(r"where's","where is",text)
    text=re.sub(r"\'ll"," will",text)
    text=re.sub(r"\'ve"," have",text)
    text=re.sub(r"\'re"," are",text)
    text=re.sub(r"\'d"," would",text)
    text=re.sub(r"won't","will not",text)
    text=re.sub(r"can't","cannot",text)
    
    #removing any special character
    text=re.sub(r"[-()\"#!@$%^&*{}?.,:]"," ",text)
    text=re.sub(r"\s+"," ",text)
    text=re.sub('[^A-Za-z0-9]+',' ', text)
    
    for word in text.split():
        if word not in stopwords:
            string+=lemma.lemmatize(word)+" "
    
    return string
    
```

```python
#cleaning the whole data
df["text"] =df.text.apply(clean_text)
```

## Feature-Extraction & Model building

```python
from sklearn.model_selection import train_test_split
```

```python
X = df["text"]
y = df["category"]

X_train,X_test,y_train,y_test = train_test_split(X,y,test_size =0.33,random_state = 42)
```

```python
#importing libraries to build a pipline
from sklearn.pipeline import Pipeline
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.svm import LinearSVC
```

```python
final_clf = Pipeline([("tfidf",TfidfVectorizer()),("clf",LinearSVC())])
final_clf.fit(X_train,y_train)
```

```python
#making prediction using the model
predictions = final_clf.predict(X_test)
```

```python
from sklearn import metrics
print(metrics.classification_report(y_test,predictions))
```

```python
#printing overall accuracy
print(metrics.accuracy_score(y_test,predictions))
```

```python
#printing confusion matrix
print(metrics.confusion_matrix(y_test,predictions))
```
